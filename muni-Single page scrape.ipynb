```python
from selenium import webdriver
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import time
```

```python
#build the main table
df = pd.DataFrame({'cusip': [], 'name': [], 'vd_content': [],'vd_date':[],'download':[]})
#disclosure web page list
file = open("2012-2018.txt")
vd = file.read()
file.close
testr=re.findall(r"E.\d{6}",vd)
root = "https://emma.msrb.org/MarketActivity/ContinuingDisclosureDetails/"
urls =  [ root + e for e in testr]
```

```python
urls_test = urls[:15]
```

```python
#loop test run
for url in urls_test:
    #click the accept button
    driver = webdriver.Chrome()
    driver.get(url)
    driver.implicitly_wait(15)
    driver.get(url)
    driver.implicitly_wait(15) 
    button = driver.find_element_by_xpath("//div/input[@type='submit']")
    button.click()
   
    #scrape the web
    time.sleep(20)
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    driver.quit

    #get cusip 6 digit info
    list = soup.find_all("a", {"class": "cdCusip6Link"})
    cusip_code=[]
    cusip_name=[]
    for i in list:
        cusip_code.append(i.text.split("(")[-1][:-1])
        cusip_name.append(i.text.split("(")[0][:-2])
    
    #get disclosure file name and submit date
    title = soup.find("table",{"class":"PPCDheaderTable"}) 
    content = [title.get_text().split(":")[-1].split(",")[0]] * len(cusip_code)
    date = [title.get_text().split(":")[-1].split(",")[-1][-13:-3]] * len(cusip_code)

    #get the download path
    section = soup.find("td",{"class":"CDDocList"})
    path = section.find("a",{"target":"_blank"}).attrs['href']
    download = [str(path)] *  len(cusip_code)
    print(download)
    
    #add info to main table
    dictt = {'cusip': cusip_code, 'name': cusip_name, 'vd_content': content,'vd_date':date, 'download': download}
    dff = pd.DataFrame(dictt)
    df = df.append(dff)
```
